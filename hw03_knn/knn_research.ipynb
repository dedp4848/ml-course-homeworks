# Исследование метода KNN
## Датасет: Iris Species

### 1. Загрузка данных и первичный анализ

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB

# Загружаем датасет
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

# Смотрим на данные
print('Размер датасета:', df.shape)
print()
df.head()
```

```python
# Информация о данных
df.info()
```

```python
# Статистики
df.describe()
```

```python
# Распределение классов
print('Распределение классов:')
df['target'].value_counts().sort_index()
```

**Выводы по анализу:**
- 150 объектов, 4 признака (все числовые)
- 3 класса (по 50 объектов каждый) — сбалансированный датасет
- Пропусков нет
- Признаки: длина/ширина чашелистика и лепестка

### 2. Подготовка данных

```python
# Разделяем признаки и целевую переменную
X = df.drop('target', axis=1)
y = df['target']

# Разделяем на train и test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print('Train:', X_train.shape)
print('Test:', X_test.shape)
```

```python
# Масштабирование признаков
# Для KNN это важно, потому что метод использует расстояния между объектами
# Если признаки в разных масштабах, то признаки с большими значениями будут доминировать

scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X.columns)
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X.columns)

print('Масштабирование выполнено')
```

### 3. Обучение KNN

```python
# Базовая модель с k=5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

# Предсказания
y_pred = knn.predict(X_test)

# Метрики
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Precision:', precision_score(y_test, y_pred, average='macro', zero_division=0))
print('Recall:', recall_score(y_test, y_pred, average='macro', zero_division=0))
print('F1:', f1_score(y_test, y_pred, average='macro', zero_division=0))
print()
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
```

### 4. Исследование влияния числа соседей (k)

```python
# Проверяем разные значения k
results = []

for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    acc = knn.score(X_test, y_test)
    results.append((k, acc))
    print(f'k={k:2d} -> Accuracy={acc:.3f}')

best_k = max(results, key=lambda x: x[1])
print(f'\nЛучшее k: {best_k[0]} (Accuracy={best_k[1]:.3f})')
```

### 5. Сравнение метрик расстояния

```python
# Проверяем разные метрики
metrics = ['euclidean', 'manhattan', 'minkowski']

for metric in metrics:
    if metric == 'minkowski':
        knn = KNeighborsClassifier(n_neighbors=5, metric=metric, p=2)
    else:
        knn = KNeighborsClassifier(n_neighbors=5, metric=metric)
    knn.fit(X_train_scaled, y_train)
    acc = knn.score(X_test, y_test)
    print(f'{metric}: Accuracy={acc:.3f}')
```

### 6. Влияние масштабирования

```python
# Без масштабирования
knn_no_scale = KNeighborsClassifier(n_neighbors=5)
knn_no_scale.fit(X_train, y_train)
acc_no_scale = knn_no_scale.score(X_test, y_test)

# Со масштабированием
knn_scaled = KNeighborsClassifier(n_neighbors=5)
knn_scaled.fit(X_train_scaled, y_train)
acc_scaled = knn_scaled.score(X_test, y_test)

print(f'Без масштабирования: {acc_no_scale:.3f}')
print(f'Со масштабированием: {acc_scaled:.3f}')
```

### 7. Подбор гиперпараметров (GridSearchCV)

```python
# Параметры для поиска
param_grid = {
    'n_neighbors': [1, 3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

# GridSearch с кросс-валидацией
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy')
grid.fit(X_train_scaled, y_train)

print('Лучшие параметры:', grid.best_params_)
print('Лучшая accuracy (CV):', grid.best_score_)
```

```python
# Проверяем лучшую модель на тесте
best_model = grid.best_estimator_
test_acc = best_model.score(X_test, y_test)
print(f'Accuracy на тестовой выборке: {test_acc:.3f}')
```

### 9. Выводы

**Результаты исследования:**

1. **Масштабирование** — важно для KNN, улучшает точность, т.к. метод основан на расстояниях

2. **Число соседей (k)** — оптимальное значение k=3-7 для этого датасета
   - Малое k → модель чувствительна к шуму
   - Большое k → модель слишком усредняет

3. **Метрика расстояния** — euclidean и manhattan дают схожие результаты на Iris

4. **Веса (weights)** — 'distance' может работать лучше, т.к. ближние соседи важнее

5. **Сравнение с другими моделями** — KNN показывает результат сравнимый с логистической регрессией и Наивным Байесом

**Ограничения KNN:**
- Медленная работа на больших данных (нужно хранить все обучающие данные)
- Чувствительность к масштабу признаков
- Чувствительность к выбросам при малом k
